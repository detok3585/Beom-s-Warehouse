```
from sklearn import neighbors, datasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression

data = datasets.load_breast_cancer()
data
```

입출력 데이터 설정

```
x = data.data
y = data.target
```

데이터 훈련셋과 테스트셋으로 나누기

```
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state = 0)
```

KNeighborsClassifier 모델 생성 후 fit.

```
k=6
KN = neighbors.KNeighborsClassifier(k,weights='distance').fit(xtrain,ytrain)
```

```
KN
```

결과 예측

```
KNResult = KN.predict(xtest)
```

모델 정확성 평가

```
metrics.accuracy_score(ytest,KNResult)
```

SVM 모델 만든 후 fit

```
SVM = SVC(kernel='linear').fit(xtrain,ytrain)
```

결과 예측

```
SVMResult = SVM.predict(xtest)
```

모델 정확성 평가

```
metrics.accuracy_score(ytest,SVMResult)
```

의사결정분류기 모델 생성후 fit

```
dstClf = DecisionTreeClassifier()
dstClf.fit(xtrain, ytrain)
```

결과 예측

```
dstClfResult = dstClf.predict(xtest)
```

모델 정확성 평가

```
metrics.accuracy_score(ytest,dstClfResult)
```

로지스틱 회귀 모델 생성 후 fit

```
log = LogisticRegression().fit(xtrain,ytrain)
```

결과 예측

```
logReslut = log.predict(xtest)
```

모델 정확성 평가

```
metrics.accuracy_score(ytest,logReslut)
```

4가지 방식 비교.
- KNeighborsClassifier
- SVM
- DecisionTree
- LogisticRegression


```
metrics.accuracy_score(ytest,KNResult),metrics.accuracy_score(ytest,SVMResult),metrics.accuracy_score(ytest,dstClfResult),metrics.accuracy_score(ytest,logReslut)
```
