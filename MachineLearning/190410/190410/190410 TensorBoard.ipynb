```
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorboardcolab import *
```

```
tbc=TensorBoardColab()
```

```
sess = tf.Session()
```

![image](https://user-images.githubusercontent.com/46669509/55843801-0da92180-5b75-11e9-8838-5c02d6a6fe90.png)


```
node1 = tf.constant(3.0,tf.float32)
node2 = tf.constant(5.0,tf.float32)
node3 = tf.add(node1,node2)
```

```
sess.run(node3)
```

![image](https://user-images.githubusercontent.com/46669509/55843822-20bbf180-5b75-11e9-94c7-a1d9f5b3f182.png)


```
a1 = tf.placeholder(tf.float32)
b1 = tf.placeholder(tf.float32)
adder_node = a1+b1
sess.run(adder_node,feed_dict={a1:3,b1:5})
sess.run(adder_node,feed_dict={a1:[3,2],b1:[5,]})
```

![image](https://user-images.githubusercontent.com/46669509/55844041-75ac3780-5b76-11e9-889a-2ceee451295c.png)

```
addTest = adder_node * 3
```

```
sess.run(addTest,feed_dict={a1:4,b1:5})
```
