```
data = pd.read_csv("/content/gdrive/My Drive/Colab Notebooks/data/Sample02-k-means.csv")
data.head()
```

```
plt.scatter(data['x'],data['y'])
plt.show()
#elbow 클러스터 영역을 구해주는 알고리즘
```

```
x1 = data['x'].values
y1 = data['y'].values
data2 = np.array(list(zip(x1,y1))).reshape(len(x1),2)
data2
data3 = []
```

![image](https://user-images.githubusercontent.com/46669509/55693490-55e80880-59ea-11e9-8f61-ca5755ee6034.png)


![image](https://user-images.githubusercontent.com/46669509/55693474-4799ec80-59ea-11e9-8925-b825fac0edd2.png)


```
K = range(1,10)
for k in K:
  kmeanModel = KMeans(n_clusters=k).fit(data2)
  d= cdist(data2, kmeanModel.cluster_centers_,'euclidean')
  d2 = sum(np.min(d,axis=1))/data2.shape[0] # 평균 - 각 거리의 최소값을 합해 전체로 나누어준 것
  data3.append(d2)

  
# 위 코드는 K-평균 알고리즘을 구현한 것
# kmeanModel.cluster_centers_ 는 클러스터 중심값
# scipy 의 cdist 함수는 두 개의 벡터 A, B 를 받아서 AxB 의 모든 페어에 대한 계산 결과를 2차원 배열로 리턴한다. 
```

```
plt.plot(K, data3, 'bx-')
plt.show()
```
